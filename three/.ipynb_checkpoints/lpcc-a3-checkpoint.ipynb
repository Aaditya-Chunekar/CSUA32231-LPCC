```python
mdt=[]
mnt=[]
ir=[]
d=1
m=0
```

```python
with open("code.asm", "r") as file:
    for line in file:
        if(m):
            mdt.append([d,line])
            d+=1
        #below is for ir ====
        elif 'MACRO' not in line:
            ir.append(line)
        #====================
        if 'MACRO' in line:
            m=1
            if ',' in line:
                mnt.append([line.split()[1],len(line.split(',')),d])
            else:
                mnt.append([line.split()[1],0,d])
        if 'MEND' in line:
            m=0
```

```python
for i in mnt:
    print(*i)
```

```python
for i in mdt:
    print(*i)
```
