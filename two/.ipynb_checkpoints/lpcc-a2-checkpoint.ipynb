```python
mdt=[]
mnt=[]
d=1
m=0
```

```python
with open("code.asm", "r") as file:
    for line in file:
        if(m):
            mnt.append([d,line])
            d+=1
            
        if 'MACRO' in line:
            m=1
            mdt.append([line.split()[1],0,d])
        if 'MEND' in line:
            m=0
```

```python
for i in mnt:
    print(*i)
```

```python
for i in mdt:
    print(*i)
```
