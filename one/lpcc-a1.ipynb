```python
with open("code.asm", "r") as file:
    for line in file:
        pretokens = line.strip().split()

        if not pretokens:
            continue

        # below separates one length & beginning symbols
        # if len(pretokens[0]) == 1:
        #     print("symbol(s):", pretokens[0])
        #     del pretokens[0]

        print("tokens:", end="")
        for token in pretokens:
            # below is for ignoring comments as tokens
            if token == ";":
                break
            print(token, end=", ")
        print('\n-----------------------')
```
